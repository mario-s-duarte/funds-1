```python
def assess_etf(etf:str):
    try:
        series = yf.Ticker(etf).history(period="max")['Close']
    except Exception as e:
        print(f"{etf}: {e}")
        return

    # Get monthy values
    resampled_series = series.resample('BM').first()

    # fit an exponencial curve to the data and print the parameter

    # Function to define the exponential curve
    def exponential_func(x, a, b):
        return b * a**x
    # Convert dates to a numerical format
    x_data = np.arange(1,len(resampled_series)+1).astype('float64')
    y_data = resampled_series.astype('float64')

    # Fit the exponential curve
    initial_guess = [1, y_data.min()]
    try:
        popt, pcov = curve_fit(exponential_func, x_data, y_data, maxfev=2000000)
    except Exception as e:
        print(f"{etf}: {e}")
        return

    # Generate the fitted curve
    y_fit = exponential_func(x_data, *popt)

    kpi = (y_data.iloc[-1]-y_fit[-1]) / y_data.iloc[-1]

    return {'popt':popt,'y_data':y_data, 'y_fit':y_fit, 'kpi': kpi}

def plot_etf(etf: str, popt, y_data, y_fit, kpi):
    # Plot the original Series and the fitted curve
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 4))

    ax1.plot(y_data.index, y_data.values, label='Original Series', color='blue')
    ax1.plot(y_data.index, y_fit, label='Fitted Exponential Curve', color='red')

    ax1.set_title(f'{etf} - {popt[1]:.2f}+{popt[0]:.6f}^x')

    ax1.set_xlabel('Date')
    ax1.set_ylabel('Value')

    ax2.plot(y_data.index, y_data - y_fit, label='Residuals', color='blue')
    ax2.axhline(0, color='red', linestyle='--', linewidth=2, label='y=0')

    ax2.set_title(f'{etf} - Residuals {kpi:.6f}')

    ax2.set_xlabel('Date')
    ax2.set_ylabel('Value')
```

```python
etf_data = {etf['Symbol']:(assess_etf(etf['Symbol']) | {'name': etf['ETF']}) for etf in etfs.values()}
pd.DataFrame(data=[(key,val['name'],val['kpi']) for key, val in etf_data.items()], columns=['Symbol','Name','KPI']).sort_values('KPI')
```

```python
for key, val in etf_data.items():
    plot_etf(key, val['popt'], val['y_data'], val['y_fit'], val['kpi'])
```

# Research All ETFs

```python
def etf_history(etf) -> pd.Series:
    symbol = etf.Symbol
    if etf.Exchange == "XPAR":
        symbol += ".PA"
    elif etf.Exchange == "XETR":
        symbol += ".DE"
    elif etf.Exchange == "XLON":
        symbol += ".L"
    elif etf.Exchange == "XAMS":
        symbol += ".AS"
    try:
        ds = yf.Ticker(symbol).history(period="6mo")['Close']
        ds.index = ds.index.date
        ds = ds.sort_index()
    except Exception:
        return None
    if etf.Currency == "USD":
        max_date = min(ds.index.max(), exchange_rates.index.max())
        min_date = max(ds.index.min(), exchange_rates.index.min())
        ds = ds[min_date:max_date] * exchange_rates[min_date:max_date]
        ds = ds.dropna()
    ds.name = symbol
    return ds
```

```python
# Get the list of tradable ETFs in BEST. Each row contains Exchange, Symbol, Currency
etfs = pd.read_csv("etf.csv")
# Create a DataFrame with the etf 6 month historical close price
df_quotes = pd.concat([x for x in [etf_history(etf) for etf in etfs.itertuples()] if x is not None],axis=1)
df_quotes
```

```python
# Get the amount of Nulls in the historical quotes
df_nan = pd.DataFrame([[etf,df_quotes[etf].isnull().sum()] for etf in df_quotes.columns],columns=['symbol','nan_amount'])
print(df_nan.groupby('nan_amount').count().reset_index())
```

```python
# Ignore ETFs with more than 5 nulls
df_quotes = df_quotes[df_nan[df_nan['nan_amount']<=5]['symbol']].sort_index()
print(len(df_quotes.columns))
```

```python
# Fill in NaN
df_quotes = df_quotes.ffill().bfill()
```

```python
from pypfopt.expected_returns import mean_historical_return, ema_historical_return
from pypfopt.risk_models import CovarianceShrinkage
from pypfopt.efficient_frontier import EfficientFrontier
from pypfopt import objective_functions
import cvxpy
import numpy as np
cvxpy.installed_solvers()
```

```python
def valid_date(d: date) -> date:
    if d in df_quotes.index:
        return d
    return max(dd for dd in df_quotes.index if dd <= d)

```

### Parameters

```python
risk_free_rate = 0.028
last_n = 58
span = 3
cov_n = 153

first_date = valid_date(df_quotes.index.max() - timedelta(days=last_n))
```

### Best Portofolio from all the funds

```python
### Inputs
# Funds to use

mu = mean_historical_return(df_quotes.loc[first_date:])
S = CovarianceShrinkage(df_quotes).ledoit_wolf()
ef = EfficientFrontier(mu, S, solver='SCS',weight_bounds=(0,1))
ef.max_sharpe(risk_free_rate=risk_free_rate)
cleaned_weights = ef.clean_weights()
cleaned_weights = [(s,w) for s,w in cleaned_weights.items() if w > 0.0]
df = pd.DataFrame(cleaned_weights,columns=['symbol','w']).sort_values(by='w',ascending=False)
df
```

```python
ef.portfolio_performance(verbose=True,risk_free_rate=risk_free_rate)
```
